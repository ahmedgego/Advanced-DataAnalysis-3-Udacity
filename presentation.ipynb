### Investigation Overview
In this investigation, I wanted to look at the relationships between duration and usertype , age of member , gender of member 
and user type. 
### Dataset Overview
The dataset contains 325107 entries with 15 features including(tripduration,starttime,stoptime,start station id,start station name,start station latitude,start station longitude,end station id,end station name,end station latitude,end station longitude,bikeid,usertype,birth year,gender)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
# load in the dataset into a pandas dataframe
df_bike= pd.read_csv('data_cleaned.csv')
```

## Distribution of months 

> The number of trips are increased in (may to oct) this months of summer in the first months of the year the trips are decreased because to winter

```python
base_color = sb.color_palette()[0]
sb.countplot(data=df_bike, x='month', color=base_color);
plt.xlabel('The trip month')
plt.title('The frequent trips in months');
```

## Distribution of age 

> people that have (25 to 35) are the most members to make trips 

```python
base_color = sb.color_palette()[0]
plt.hist(data=df_bike, x='member_age', color=base_color);
plt.xlabel('The age of members')
plt.title('The frequent of age for members');
```

## Distribution of gender

> males have the largest proportion then females come

```python
gender_counts = df_bike['gender'].value_counts()
plt.figure(figsize=[8, 5])
plt.pie(gender_counts, labels = gender_counts.index, autopct='%1.0f%%', startangle=90, counterclock = False);
plt.title("The distribution of members gender ");
```

## Distribution of user type 

> the subscribers have larger proportion then the customers

```python
user_counts = df_bike['usertype'].value_counts()
plt.figure(figsize=[8, 5])
plt.pie(user_counts, labels = user_counts.index, autopct='%1.0f%%', startangle=90, counterclock = False);
plt.title("The distribution of members type ");
```

## Relationship between (usertype & gender & months)


```python
g = sb.catplot(data=df_bike, x='month', col="usertype", hue='gender', kind='count', sharey = False)
g.set_axis_labels("Day", "Number of Trips")
g.fig.suptitle('Relationship Between User Type and Gender, based on month',fontsize=14,y=1.1);
```

## Relationship between (usertype & gender & days)


```python
g = sb.catplot(data=df_bike, x='day', col="usertype", hue='gender', kind='count', sharey = False)
g.set_axis_labels("Day", "Number of Trips")
g.fig.suptitle('Relationship Between User Type and Gender, based on day',fontsize=14,y=1.1);
```
